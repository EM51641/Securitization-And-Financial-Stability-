# 2008 Global Financial Crisis Explication

# This code aims to provide reasons that pushes bank to expand their balance sheet through mortgage loans 

For a better understanding of the paper,the code refers to the paper of Hyun Song Shin available at this link:

https://www.jstor.org/stable/20485321?seq=1

```python
import numpy as np
import pandas as pd
n = 5000
T = (0.20 - 0) * np.random.random_sample((n, n)) + 0 #Banks have equity between 5% and 20% of their total landing program 
dct = {'Y(claims from households)_in_$':(np.random.rand(1,n).tolist())[0],'Bank backed security_in_$':((T).tolist())[0] }
```

```python
B=pd.DataFrame(dct)*5*10**12
```

```python
B
```

```python
a = (0.005/n - 0) * np.random.random_sample((n, n)) + 0 #Banks own typically between (0.005/10000)% to 0% of other shares
row,col = np.diag_indices(a.shape[0])
a[row,col] = np.array(n*[0])
array = a
```

```python
shared_Mortgages = pd.DataFrame(array)
shared_Mortgages
```

```python
B['Shared Equity Amount_in_$'] = 0
for x in range(n):
    B.iloc[x,-1] = np.dot(shared_Mortgages.iloc[x].values,B.iloc[:,1].values)
```

```python
B['Residuals_in_$'] = B['Y(claims from households)_in_$'] + B['Shared Equity Amount_in_$']- B['Bank backed security_in_$']
```

```python
#Some residuals may be negative.it means that the bank actually owns more mortgage claims than it issued,then it can be 
#accounted as an asset
```

```python
B
```

```python
correlation_between_defaults = (0.10 - 0.001) * np.random.random_sample((n, 1)) + 0.001
Probability_of_defaults = (0.03 - 0.001) * np.random.random_sample((n, 1)) + 0.001
```

```python
B['Correlation_between_defaults'] = correlation_between_defaults
B['Probability_of_defaults'] = Probability_of_defaults
```

```python
from scipy.stats import norm                                                                              
```

```python
rd = pd.DataFrame(np.random.rand(1,n))
```

```python
Repayment_probability = B['Y(claims from households)_in_$'].values*norm.cdf((rd*np.sqrt(B['Correlation_between_defaults'])-norm.ppf(B['Probability_of_defaults']))/np.sqrt(1-B['Correlation_between_defaults']))
```

```python
B['Repayement Total_in_$'] = pd.DataFrame(Repayment_probability).T
```

```python
repayment_threshold_2008crisis = 0.95 #* B['Y(claims from households)']

repayment_threshold_2008crisis_Max = norm.cdf((norm.ppf(repayment_threshold_2008crisis)*np.sqrt(1-B['Correlation_between_defaults'])+norm.ppf(B['Probability_of_defaults']))/(np.sqrt(B['Correlation_between_defaults'])))
```

```python
B
```

```python
B['Leverage'] = np.maximum((B['Shared Equity Amount_in_$']*0.995 + B['Repayement Total_in_$'])/ B['Residuals_in_$'],0)
```

```python
#Proportion of bank's i debts detained by outside shareholders
Z = np.dot([np.identity(n)-np.array(shared_Mortgages)],np.ones((n,1)))
B['Outsided claimholders']=pd.DataFrame(Z.tolist()[0])
```

```python
Leverage_Agregate_sector = (B['Repayement Total_in_$']+B['Shared Equity Amount_in_$']).sum()/(B['Residuals_in_$'].sum()) #Agregate leverage 
```
